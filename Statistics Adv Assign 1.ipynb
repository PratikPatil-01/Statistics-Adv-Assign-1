The Probability Density Function (PDF) is a function that describes the probability distribution of a continuous random variable. It gives the probability that the variable will take on a particular value within a given range.
</br>
Properties of the probability density function:
</br>
The probability density function is always non-negative: 
The probability that the random variable falls within any particular interval is given by the integral of the probability density function over that interval.</br>
The probability density function is a fundamental concept in probability theory and statistics. It is used to describe and analyze the behavior of continuous random variables, such as heights, weights, and temperatures. Examples of probability density functions include the normal distribution, uniform distribution, exponential distribution, and many others.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data points for x-axis
x = np.linspace(-5, 5, 1000)

# Define parameters for normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Calculate the PDF for each data point
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='PDF of Normal Distribution', color='blue')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

def normal_pdf(x, mu, sigma):
    """
    Calculate the probability density function (PDF) of a normal distribution
    at a given point.

    Parameters:
        x (float): The point at which to evaluate the PDF.
        mu (float): The mean of the normal distribution.
        sigma (float): The standard deviation of the normal distribution.

    Returns:
        float: The value of the PDF at the given point.
    """
    pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))
    return pdf

# Example usage:
mean = 0
std_dev = 1
point = 1.5
pdf_value = normal_pdf(point, mean, std_dev)
print("PDF at point", point, ":", pdf_value)
```

```python
import numpy as np

def poisson_cdf(k, lambd):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution
    with given mean (lambda) at a given point (k).

    Parameters:
        k (int): The point at which to evaluate the CDF.
        lambd (float): The mean of the Poisson distribution.

    Returns:
        float: The value of the CDF at the given point.
    """
    cdf = np.exp(-lambd) * np.sum([(lambd**i) / np.math.factorial(i) for i in range(k + 1)])
    return cdf

# Example usage:
mean = 2
point = 3
cdf_value = poisson_cdf(point, mean)
print("CDF at point", point, ":", cdf_value)
```

The Binomial and Poisson distributions are both used to model the number of occurrences of events, but they have different characteristics and applications. Here are the main differences between the two distributions:
</br>
Nature of Events:
</br>
Binomial Distribution: The Binomial distribution is used to model the number of successes in a fixed number of independent Bernoulli trials, where each trial has two possible outcomes (success or failure). The trials are discrete and independent.
Poisson Distribution: The Poisson distribution is used to model the number of events occurring in a fixed interval of time or space, where the events are independent and occur at a constant average rate.
</br>
Number of Trials:
</br>
Binomial Distribution: The Binomial distribution requires a fixed number of trials or experiments (n).
</br>
Poisson Distribution: The Poisson distribution does not require a fixed number of trials. It is often used when the number of trials is large (n) and the probability of success (p) is small.
</br>
Probability of Success:
Binomial Distribution: The Binomial distribution requires a constant probability of success (p) for each trial.
Poisson Distribution: The Poisson distribution does not require a constant probability of success. It is characterized by a single parameter (), which represents the average rate of occurrence of events.
</br>
Parameters:
Binomial Distribution: The Binomial distribution is characterized by two parameters: the number of trials (n) and the probability of success (p).
Poisson Distribution: The Poisson distribution is characterized by a single parameter (), which represents the average rate of occurrence of events.
</br>
Shape:
Binomial Distribution: The Binomial distribution is bounded, with a finite number of possible outcomes.
Poisson Distribution: The Poisson distribution is unbounded, with an infinite number of possible outcomes.

In a normal distribution, the least frequent data points appear farthest from the mean.
</br>
Since the normal distribution is symmetric around its mean, the data points become less frequent as they move away from the mean in both directions towards the tails of the distribution. Therefore, the least frequent data points appear at the extremes, or tails, of the distribution.
</br>
For a standard normal distribution (mean = 0, standard deviation = 1), the least frequent data points are located at the tails, which are the values furthest from the mean, specifically at the ends of the distribution where the data is less likely to occur.
</br>
In summary, in a normal distribution, the least frequent data points appear farther from the mean, towards the tails of the distribution.
